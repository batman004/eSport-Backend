```python
def viz(name):
    

#     fig, ax1 = plt.subplots(figsize=(23,8))
#     # team_name = str(widgets.Text.value)
#     #print(country_name)
#     Bteam= df[df['bat_team']==name]
#     # team
#     total_by_team = Bteam['total']
#     # total_by_team
#     ax1=sns.distplot(total_by_team)
#     ax1.set_title(name + ": Total Runs Scored")
#     ax1.set(xlabel="Total Runs Scored", ylabel =name)

#     print()
    fig, ax = plt.subplots(figsize=(23,8))
    BBteam= df[df['bowl_team']==name]
    wicket_by_team = BBteam['wickets']
    ax.scatter(wicket_by_team,BBteam['overs'])
    ax.set_xlabel('wickets')
    ax.set_ylabel('Overs')
    ax.set_title(name + ": Wickets vs Overs")
    plt.show()
    
bat = list(df.bat_team.unique())
bat
bowl = list(df.bowl_team.unique())
final_list = list(set(bat) | set(bowl))
final_list.sort()
final_list

for country in final_list:
    viz(country)
```
